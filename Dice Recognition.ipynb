```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/0.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image, cmap = 'gray')
```

```python
img = cv2.imread('images/0.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap = 'gray')
```

```python
img_edge = cv2.Canny(img, 10, 255)

plt.imshow(img_edge, cmap = 'gray')
```

```python
kernel = np.ones((20, 20), np.uint8)

closing = cv2.morphologyEx(img_edge, cv2.MORPH_CLOSE, kernel)

plt.imshow(closing, cmap= 'gray')
```

```python
# contours, hierarchy = cv2.findContours(closing, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours, hierarchy = cv2.findContours(closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# print(len(contours))

x, y, w, h = cv2.boundingRect(contours[0])
cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 255), 6)
    
plt.imshow(image)
```

```python
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1, 1, param1=70, param2 = 30, minRadius=1, maxRadius= 20)
# print(circles)
```

```python
circles = circles[0, :]
# print(circles)

for circle in circles:
    cv2.circle(image, (int(circle[0]), int(circle[1])), int(circle[2]), (255, 0, 255), 6)
    
plt.imshow(image)
```

```python
count_circles = len(circles)

cv2.putText(image, f'num: {count_circles}', (x+30, y-10), cv2.FONT_HERSHEY_SIMPLEX, 1,(255, 0, 0), 3)

plt.imshow(image)
```

```python
cv2.rectangle(image, (x1, y1), (x1+w1, y1+h1), (0, 255, 255), 4)
cv2.rectangle(image, (x2, y2), (x2+w2, y2+h2), (0, 255, 255), 4)

plt.imshow(image)
```

```python
circles = cv2.HoughCircles(img_edge, cv2.HOUGH_GRADIENT, 1, 10, param1=70, param2 = 25, minRadius=1, maxRadius= 40)
# print(circles)
```

```python
circles = circles[0, :]

for circle in circles:
    cv2.circle(image, (int(circle[0]), int(circle[1])), int(circle[2]), (0,0,255), 2)
    
plt.imshow(image)
```

```python
circles1 = cv2.HoughCircles(dice1, cv2.HOUGH_GRADIENT, 1, 10, param1=70, param2 = 25, minRadius=1, maxRadius= 40)
circles1 = circles1[0, :]
# print(len(circles1))

circles2 = cv2.HoughCircles(dice2, cv2.HOUGH_GRADIENT, 1, 10, param1=70, param2 = 25, minRadius=1, maxRadius= 40)
circles2 = circles2[0, :]
# print(len(circles2))

cv2.putText(image, f'num: {len(circles1)}', (x1+10, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 1,(255, 0, 255), 2)
cv2.putText(image, f'num: {len(circles2)}', (x2+10, y2-10), cv2.FONT_HERSHEY_SIMPLEX, 1,(255, 0, 255), 2)

plt.imshow(image)
```
